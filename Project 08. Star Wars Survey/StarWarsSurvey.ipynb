# Star Wars Survey
The team at FiveThirtyEight  wondered: does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?  They surveyed Star Wars fans using the online tool SurveyMonkey. They received 835 total responses, which you can download from their [GitHub repository.](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey)

For this project, we'll be cleaning and exploring the star wars survey dataset 

### 1. Overview

The data has several columns, including:

* RespondentID - An anonymized ID for the respondent (person taking the survey)
* Gender - The respondent's gender
* Age - The respondent's age
* Household Income - The respondent's income
* Education - The respondent's education level
* Location (Census Region) - The respondent's location
* Have you seen any of the 6 films in the Star Wars franchise? - Has a Yes or No response
* Do you consider yourself to be a fan of the Star Wars film franchise? - Has a Yes or No response

```python
# Read the data set into a dataframe and explore data
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head(5)
```

```python
# Review the column names
star_wars.columns
```

```python
# Remove any rows where RespondentID is NaN.
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars.head(5)
```

### 2.  Cleaning and Mapping Yes/No Columns

Convert following 2 colums to Boolean type
1. "Have you seen any of the 6 films in the Star Wars franchise?"
2. "Do you consider yourself to be a fan of the Star Wars film franchise?"


```python
# replace Yes/No with True/False

yes_no = {
    "Yes": True,
    "No": False
}
for col in [
    'Have you seen any of the 6 films in the Star Wars franchise?',
    'Do you consider yourself to be a fan of the Star Wars film franchise?']:
    star_wars[col]= star_wars[col].map(yes_no)
star_wars.head()
```

### 3. Clean and Mapping checkbox Columns

1. Convert columes ("Unnamed: 4-8" )so that it only contains the values True and False
2. Remame columes ("Unnamed: 4-8" ) to Seen_2-6, which are more intuitive.


```python
import numpy as np

movie_map = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.NaN: False
}
for col in star_wars.columns[3:9]:
    star_wars[col] =star_wars[col].map(movie_map) 

col_map = {
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
     "Unnamed: 5": "seen_3",
     "Unnamed: 6": "seen_4",
     "Unnamed: 7": "seen_5",
     "Unnamed: 8": "seen_6"
}  
star_wars = star_wars.rename(columns = col_map)

star_wars.head()
```

### 4. Clean the Ranking Columns

 * Convert ranking columns to float type
 * Give each ranking column a more descriptive name.

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
rank_map = {
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
}
star_wars = star_wars.rename(columns = rank_map)
star_wars.head()
```

### 5. Finding the Hightest-Ranked Movie

* compute the mean of each of the ranking columns
* Make a bar chart of each mean rank. 


```python
mean_rank =star_wars[star_wars.columns[9:15]].mean()

%matplotlib inline
import matplotlib.pyplot as plt
plt.bar(range(6), mean_rank)
```

Episode I, I, III rank heigher

### 6. Find the Most Viewed Moive

 * compute the sum of each of the seen columns.
 * Make a bar chart of sum of the seen columns. 


```python
total_view =star_wars[star_wars.columns[3:9]].sum()    
plt.bar(range(6), total_view)
```

Episode I, V, VI have higher view count

### 7. Exploreing the Data by Gender

* We will split a dataframe into two groups based on a binary column like 'Gender' column
* Then we will redo the two previous analyses (find the most viewed movie and the highest-ranked movie) separately for each group, and then compare the results.

```python
# Split the data into two groups based on gendere
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
# compute the mean of ranking for male and female
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.title("Male ranking")
plt.show()
plt.bar(range(6), females[females.columns[9:15]].mean())
plt.title("Female ranking")
plt.show()
```

```python
# compute the mean of view count for male and female
plt.bar(range(6), males[males.columns[3:9]].mean())
plt.title("Male view count")
plt.show()
plt.bar(range(6), females[females.columns[3:9]].mean())
plt.title("Female view count")
plt.show()
```

Male/Female differences in favorite Star Wars movie and most seen movie
- more males watches episodes 1-3
- The least of females watches Episode 3, but the ranking of Episode 3 is the highest among females
